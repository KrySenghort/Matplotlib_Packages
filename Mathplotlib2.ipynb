```python
#initialize data
np.random.seed(1)
x = np.linspace(0,8,16)
y1 = 3 + (0.5)*x + np.random.uniform(0,0.5,len(x)) # is the curve of above area
y2 = 1 + (0.25)*x + np.random.uniform(0,0.5,len(x)) # is the curve of below area

# plot data 
fig,ax = plt.subplots(figsize=(6, 6))

ax.fill_between(x,y1,y2,alpha=0.5,linewidth=0,label='area bounded by y1,y2') 
# show the filled area that are bounded by two curve above.

ax.plot(x,(y1+y2)/2,linewidth=2,label='x,(y1+y2)/2',color='r')     
# show the line curve that displaying between areas filled.

ax.set(xlim=(0,8),xticks=np.arange(1,8),ylim=(0,8),yticks=np.arange(1,8)) 
# set the limitation value of x.

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
x = np.arange(0.1, 4, 0.1)
y1 = np.exp(-x)
y2 = np.exp(-x/2)

# example variable error bar values
y1err = 0.1 + 0.1 * np.sqrt(x)   # error of y1 = 0,1+10%sqrt(x)
y2err = 0.1 + 0.1 * np.sqrt(x/2) # error of y2 = 0.1+10%sqrt(x/2)

fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3, sharex=True,figsize=(12, 6))

ax0.set_title('all errorbars')
ax0.errorbar(x, y1, yerr=y1err)
ax0.errorbar(x, y2, yerr=y2err)

ax1.set_title('only every 6th errorbar')
ax1.errorbar(x, y1, yerr=y1err, errorevery=6)
ax1.errorbar(x, y2, yerr=y2err, errorevery=6)

ax2.set_title('second series shifted by 3')
ax2.errorbar(x, y1, yerr=y1err, errorevery=(0, 6))
ax2.errorbar(x, y2, yerr=y2err, errorevery=(3, 6))

fig.suptitle('Errorbar subsampling')
plt.show()
print('blue is y1 and red is y2')
```

```python
bins = np.arange(14)
centers = bins[:-1] + np.diff(bins) / 2
y = np.sin(centers / 2)

plt.figure(figsize=(5,3),dpi=100)
plt.step(bins[:-1], y, where='post', label='step(where="post")')
plt.plot(bins[:-1], y, 'o--', color='grey', alpha=0.3)

plt.stairs(y - 1, bins, baseline=None, label='stairs()')
plt.plot(centers, y - 1, 'o--', color='grey', alpha=0.3)
plt.plot(np.repeat(bins, 2), np.hstack([y[0], np.repeat(y, 2), y[-1]]) - 1,'o', color='red', alpha=0.2)

plt.legend()
plt.title('step() vs. stairs()')
plt.show()
```

```python
# step()   :  focus mainly about the begining point of stair from one to one
# stairs() :  focus mainly about the middle point of stair from one to one. it is a frequency polygon
```

```python
# from mathplotlib.patches import Polygon
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t) # f(t) = exp(-t)cos(2*pi*t)
t1 = np.arange(0,3,0.01)

plt.figure(figsize=(5,5),dpi=70)
ax1 = plt.subplot(212)
ax1.margins(0.05)  # Default margin =  0.05, value 0 means fit
ax1.set_title('Whole picture')
ax1.plot(t1,f(t1))

ax2 = plt.subplot(221)
ax2.margins(2,2)
ax2.set_title('Zoomed out')
ax2.plot(t1,f(t1))

ax3 = plt.subplot(222)
ax3.margins(x = 0, y = -0.25)
ax3.set_title('Zoomed in')
ax3.plot(t1,f(t1))

plt.show()
# ax1 is whole picture
# ax2 is zoomed out
# ax3 is zoomed in
```
